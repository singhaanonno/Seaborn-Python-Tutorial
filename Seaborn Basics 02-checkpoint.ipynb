```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set(style='darkgrid')
```

```python
dots=sns.load_dataset('dots').query("align=='dots'")
dots.head()
```

```python
sns.relplot(x='time',y='firing_rate',data=dots)
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line')
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',hue='coherence')
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',hue='coherence',style='choice')
```

```python
palette=sns.cubehelix_palette(light=0.5,n_colors=6)
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',hue='coherence',style='choice',palette=palette)
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',size='coherence',style='choice')
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',size='coherence',style='choice',sizes=(1,5))
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',size='coherence',hue='coherence',style='choice',sizes=(1,3))
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',size='choice',hue='coherence',style='choice',sizes=(1,3))
```

# New Dataframe

```python
df=pd.DataFrame(dict(time=pd.date_range('2019-06-02',periods=500),value=np.random.randn(500).cumsum()))
df.head()
```

```python
sns.relplot(x='time',y='value',kind='line',data=df)
```

```python
g=sns.relplot(x='time',y='value',kind='line',data=df)
g.fig.autofmt_xdate()
```

# Tips dataset

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',col='time',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',row='size',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',col='size',data=tips,col_wrap=3)
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='timepoint',y='signal',hue='subject',col='region',
            row='event',height=3,kind='line',estimator=None,data=fmri)
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',style='event',data=fmri,markers=True,ci=68,err_style='bars')
```
