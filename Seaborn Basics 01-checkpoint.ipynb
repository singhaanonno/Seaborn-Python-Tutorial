```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
fmri['timepoint'].value_counts()
```

```python
ten = fmri.where(fmri['timepoint'] == 10)
ten=ten.dropna()
ten.head()
```

```python
sum(ten['signal'])/len(ten['signal']) # Here a is the mean value of timepoint 18
```

```python
max(ten['signal'])
```

```python
min(ten['signal'])
```

```python
eighteen = fmri.where(fmri['timepoint'] == 18)
eighteen=eighteen.dropna() #to avoid all the nun values
eighteen.head()
```

```python
a=sum(eighteen['signal'])/len(eighteen['signal'])# Here a is the mean value of timepoint 18
print(a)
b=eighteen['signal']
a=(a-b)**2
c=(sum(a)/len(a))**(1/2)
c
# Here c is the standard deviation of eighteen
# light gray shadow is between (a+c) and (a-c)
# standard deviation was in our inter syllabus
# To know more watch the youtube videos of standard deviation and normal distribution 
```

```python
max(eighteen['signal'])
# Its unnecessary
```
