```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set(style='darkgrid')
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

# Catplot

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(y='day',x='total_bill',data=tips)
#We have rotated the axis dude
```

```python
sns.catplot(x='day',y='total_bill',data=tips,jitter=False)
#Here we provided a jitter argument
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='sex')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='size')
```

```python
sns.catplot(x='smoker',y='tip',data=tips)
```

```python
#now if we want to change the order we can pass order argument

sns.catplot(x='smoker',y='tip',data=tips,order=['No','Yes'])
```

# boxplot

![](boxplot.png)

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips,hue='sex')
```

# Boxen Plot

```python
sns.catplot(x='day',y='total_bill',kind='boxen',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='boxen',hue='sex',data=tips)
```

# Violin Plot

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',data=tips)
```

```python
sns.catplot(y='day',x='total_bill',kind='violin',hue='sex',data=tips)
```

```python
sns.catplot(y='day',x='total_bill',kind='violin',hue='sex',data=tips,split=True)
```

```python
sns.catplot(y='day',x='total_bill',kind='violin',hue='sex',data=tips,split=True,inner='stick')
```

```python
g=sns.catplot(x='day',y='total_bill',kind='violin',data=tips,inner=None)
sns.swarmplot(x='day',y='total_bill',color='k',size=3,data=tips,ax=g.ax)
```
