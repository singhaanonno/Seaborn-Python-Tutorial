# Visualizing Distribution of The Data 
## distplot()
## kdeplot()
## jointplot()
## rugplot()

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set(style='darkgrid')
```

```python
x=np.random.randn(100)
sns.distplot(x)
```

```python
sns.distplot(x,kde=False)
```

```python
sns.distplot(x,kde=False,rug=True)
```

```python
sns.distplot(x,kde=False,rug=True,bins=40)
```

```python
sns.distplot(x,kde=True,hist=False,rug=True,bins=40)
```

```python
sns.kdeplot(x,shade=True)
```

```python
sns.kdeplot(x,shade=True,bw=0.2)
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
x=tips['total_bill']
y=tips['tip']
```

```python
sns.jointplot(x=x,y=y)
```

```python
sns.jointplot(x=x,y=y,kind='hex')
```

```python
sns.jointplot(x=x,y=y,kind='kde')
```

```python
f,ax=plt.subplots(figsize=(6,6))
cmap=sns.cubehelix_palette(as_cmap=True,dark=0,light=1,reverse=True)
sns.kdeplot(x,y,cmap=cmap)
```

```python
f,ax=plt.subplots(figsize=(6,6))
cmap=sns.cubehelix_palette(as_cmap=True,dark=0,light=1,reverse=True)
sns.kdeplot(x,y,cmap=cmap,n_levels=60,shade=True)
```

```python
g=sns.jointplot(x,y,kind='kde',color='k')
g.plot_joint(plt.scatter,c='w',s=30,linewidth=1,marker='o')
g.ax_joint.collections[0].set_alpha(0)
```

```python
g=sns.jointplot(x,y,kind='kde',color='k')
g.plot_joint(plt.scatter)
```
